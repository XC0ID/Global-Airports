**Goal:** Understand the structure, distributions, and relationships in the airports dataset.

## 1. Imports & Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 12
sns.set_style('whitegrid')

print('Libraries loaded ✔')
```

## 2. Load Data

```python
df = pd.read_csv('../data/airports.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 3. Basic Info

```python
print('=== Data Types ===')
print(df.dtypes)
print('\n=== Missing Values ===')
print(df.isnull().sum())
print('\n=== Unique Values per Column ===')
print(df.nunique())
```

```python
df.describe().round(2)
```

## 4. Target Distribution — Hub Type

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df['hub_type'].value_counts().plot(kind='bar', ax=axes[0], color=['steelblue','coral','green'], edgecolor='black')
axes[0].set_title('Hub Type — Count')
axes[0].set_xlabel('Hub Type')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=0)

df['hub_type'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%',
                                    colors=['steelblue','coral','green'], startangle=90)
axes[1].set_title('Hub Type — Proportion')
axes[1].set_ylabel('')

plt.suptitle('Distribution of Airport Hub Types', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5. Passenger Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(df['passengers_millions'], bins=15, color='steelblue', edgecolor='black', alpha=0.8)
axes[0].set_title('Passenger Volume Distribution')
axes[0].set_xlabel('Passengers (millions)')
axes[0].set_ylabel('Count')

df.boxplot(column='passengers_millions', by='hub_type', ax=axes[1], patch_artist=True)
axes[1].set_title('Passengers by Hub Type')
axes[1].set_xlabel('Hub Type')
axes[1].set_ylabel('Passengers (millions)')
plt.suptitle('')

plt.tight_layout()
plt.show()
```

## 6. Top 15 Busiest Airports

```python
top15 = df.nlargest(15, 'passengers_millions')[['name', 'country', 'passengers_millions', 'hub_type']]

plt.figure(figsize=(11, 6))
colors = ['#e74c3c' if h == 'International' else '#3498db' for h in top15['hub_type']]
plt.barh(top15['name'], top15['passengers_millions'], color=colors, edgecolor='black', alpha=0.85)
plt.xlabel('Passengers (millions)')
plt.title('Top 15 Busiest Airports', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

top15
```

## 7. Airports by Country

```python
country_counts = df['country'].value_counts().head(12)

plt.figure(figsize=(11, 5))
country_counts.plot(kind='bar', color='steelblue', edgecolor='black', alpha=0.85)
plt.title('Number of Airports per Country (Top 12)', fontsize=13)
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 8. Correlation Heatmap

```python
num_cols = ['latitude', 'longitude', 'altitude_ft', 'passengers_millions', 'runways']
corr = df[num_cols].corr()

plt.figure(figsize=(8, 6))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix — Numeric Features', fontsize=13)
plt.tight_layout()
plt.show()
```

## 9. Altitude vs Passengers Scatter

```python
palette = {'International': '#e74c3c', 'Regional': '#3498db', 'Domestic': '#2ecc71'}

plt.figure(figsize=(10, 6))
for hub, color in palette.items():
    mask = df['hub_type'] == hub
    plt.scatter(df.loc[mask, 'altitude_ft'], df.loc[mask, 'passengers_millions'],
                label=hub, color=color, s=80, alpha=0.75, edgecolors='black', linewidth=0.4)

plt.xlabel('Altitude (ft)')
plt.ylabel('Passengers (millions)')
plt.title('Altitude vs Passenger Volume', fontsize=13)
plt.legend(title='Hub Type')
plt.tight_layout()
plt.show()
```

## 10. Geographic Distribution

```python
plt.figure(figsize=(14, 7))
ax = plt.gca()
ax.set_facecolor('#e8f4f8')

for hub, color in palette.items():
    mask = df['hub_type'] == hub
    sizes = df.loc[mask, 'passengers_millions'] * 3
    plt.scatter(df.loc[mask, 'longitude'], df.loc[mask, 'latitude'],
                c=color, s=sizes, label=hub, alpha=0.8, edgecolors='black', linewidth=0.4)

plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')
plt.title('Global Airport Map  (bubble size = passengers)', fontsize=13)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Hub Type')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.tight_layout()
plt.show()
```
